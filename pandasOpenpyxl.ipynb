```python
import pandas as pd

# Alterar valor de uma coluna num Excel 

produtos_df = pd.read_excel('Produtos.xlsx')
#pd.DataFrame(produtos_df) OU display()
display(produtos_df)

# Altera valor de umca coluna do DataFrame em memória consoante um valor de outra coluna
# tabela.loc.[tabela['coluna'] == 'conteudo linha', 'coluna a alterar'] = valor
produtos_df.loc[produtos_df['Tipo'] == 'Serviço', 'Multiplicador Imposto'] = 1.5
```

```python
# Fazer calculos em DataFrame em memória
produtos_df['Preço Base Final'] = produtos_df['Preço Base Original'] * produtos_df['Multiplicador Imposto']

display(produtos_df)
```

```python
# Guardar DataFrame de memória em ficheiro Excel
produtos_df.to_excel('ProdutosMod.xlsx', index=False)
```

```python
# Alterar células ficheiros Excel
from openpyxl import Workbook, load_workbook
produtos2 = load_workbook('Produtos.xlsx')
aba_ativa = produtos2.active

for celula in aba_ativa['C']:
    if celula.value == 'Serviço':
        num_celula = celula.row
        aba_ativa[f'D{num_celula}'] = 1.5

# Guardar Ficheiro em novo Excel
produtos2.save('ProdutosOpenPYXL.xlsx')
```
